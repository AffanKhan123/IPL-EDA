### 1)  What is the count of matches played in each season?

### 2) How many runs were scored in each season?

```python
matches = {}

for i in df['id']:                        # Making a dictionary for all the matches 
    if i not in matches:
        matches[i]=0
```

```python
for i in matches:
    runs = 0
    for j in df.values:              # runs for every match
        if i == j[0]:
            runs += j[9] 
    matches[i]=runs
```

```python
season_total_runs = {}

for i in seasons.keys():
    season_total_runs[i] = 0              # total runs in every season

print(season_total_runs)
```

```python
for i in seasons.keys():
    runs = 0
    for j in df_.values:
        if i in j[2]:
            if j[0] in matches.keys():
                runs += matches[j[0]]
    
    season_total_runs[i] = runs

season_total_runs
```

```python
x = season_total_runs.keys()
y = season_total_runs.values()

fig = plt.figure(figsize = (10,5))

plt.bar(x,y, color = 'pink', width =0.4)

plt.xlabel('Seasons')
plt.ylabel('Runs')
plt.title('Total runs in every season')

plt.show()
```

### 23) Who has the highest strike rate

```python
batsman_stats = {}

for index, row in df.iterrows():
    batsman = row['batsman']
    runs = row['batsman_runs']
    balls = row['ball']
   
    if batsman in batsman_stats:
       
        batsman_stats[batsman]['runs'] += runs
        batsman_stats[batsman]['balls'] += balls
    else:
       
        batsman_stats[batsman] = {'runs': runs, 'balls': balls}


for batsman, stats in batsman_stats.items():
    runs = stats['runs']
    balls = stats['balls']
    strike_rate = (runs / balls) * 100

    
batsman_stats[batsman]['strike_rate'] = strike_rate
sorted_batsman_stats = sorted(batsman_stats.items(), key=lambda x: x[1]['strike_rate'], reverse=True)


highest_strike_rate_batsman = sorted_batsman_stats[0][0]
highest_strike_rate = sorted_batsman_stats[0][1]['strike_rate']

print(f"The batsman with the highest strike rate: {highest_strike_rate_batsman}")
print(f"Highest strike rate: {highest_strike_rate:.2f}")
```

### 25) Which stadium hosted the most no of matches322

```python
match_counts = {}

for index, row in df_.iterrows():
    stadium = row['venue']

    if stadium in match_counts:      
        match_counts[stadium] += 1
        
    else:
        match_counts[stadium] = 1


sorted_match_counts = sorted(match_counts.items(), key=lambda x: x[1], reverse=True)
most_matches_stadium = sorted_match_counts[0][0]
most_matches = sorted_match_counts[0][1]


print(f"The stadium that hosted the most matches: {most_matches_stadium}")
print(f"Total matches hosted: {most_matches}")
```

```python
df_match_counts = pd.DataFrame.from_dict(match_counts, orient='index', columns=['Match Count'])
df_match_counts.sort_values(by='Match Count', ascending=False, inplace=True)

plt.figure(figsize=(12, 6))
plt.bar(df_match_counts.index, df_match_counts['Match Count'])
plt.xticks(rotation=90)
plt.xlabel('Stadium')
plt.show()
```
